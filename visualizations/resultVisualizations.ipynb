```python
#author: ML Tlachac
#for DepreST-CAT, 2021

import pandas as pd
import matplotlib.pyplot as plt
```

```python
array = [[],[],[],[],[]]
days = ['5','6','7','8','9','10']

models = ["kNN", "LR", "SVC", "RF", "XG"]
label = "phq9"
week = 2
for split in days:
    df = pd.read_csv("results/resultsCAT" + str(week) + "week" + label + "split" + str(split) + ".csv").fillna(0)
    for m in range(0, len(models)):
        d = df[df.model == models[m]]
        d1 = d[d.nFeatures == 1]
        array[m].append(sum(d1.F1)/100)
        

plt.figure(figsize = (6, 4))
plt.plot(days, array[0], c = 'r', marker = '.', label = models[0], linewidth=2, markersize=15)
plt.plot(days, array[1], c = 'b', marker = '^', label = models[1], linewidth=2, markersize=12)
plt.plot(days, array[2], c = 'g', marker = 's', label = models[2], linewidth=2, markersize=12)
plt.plot(days, array[3], c = 'k', marker = 'X', label = models[3], linewidth=2, markersize=12)
plt.plot(days, array[4], c = 'c', marker = 'D', label = models[4], linewidth=2, markersize=12)

plt.ylim(0.35, 0.85)
plt.xlabel("PHQ-9 Cutoff", fontsize = 13.5)
plt.ylabel("Average F1", fontsize = 13.5)
plt.title("Depression Screening at Different Cutoffs", fontsize = 15)
plt.legend()

plt.savefig("resultPlots/cutoff" + label + ".png", bbox_inches = "tight")

plt.show()
plt.close()
```
