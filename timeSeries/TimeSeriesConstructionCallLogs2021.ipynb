```python
start = pd.read_csv("surveysDepreST-CAT.csv")
start.head()
```

```python
from datetime import datetime, timedelta
from dateutil import parser

year = []
month = []
day = []
for i in range(0, start.shape[0]):
    string = start.Timestamp[i]
    year.append(string[1:5])
    month.append(string[6:8])
    day.append(string[9:11])
    
df = pd.DataFrame({'year': year, 'month': month, 'day': day})
df = pd.DataFrame(pd.to_datetime(df))
df["id"] = start["id"]
df.columns = ["timestamp","id"]
df.head()
```

```python
stamps = []
for i in range(0, data.shape[0]):
    ts = int(data.UnixTimestamp[i])/1000
    stamps.append(datetime.utcfromtimestamp(ts).strftime('%Y-%m-%d'))
dataTS = pd.to_datetime(stamps, format='%Y-%m-%d', errors='ignore')
data["timestamp"] = dataTS
data.head()
```

```python
#diff = (data.timestamp[0] - data.timestamp[1]).days

array1 = []
array2 = []
for i in range(0, df.shape[0]):
    values1 = []
    values2 = []
    temp = data[data["id"] == df["id"][i]]
    temp1 = temp[temp.Direction != 2].reset_index()
    temp2 = temp[temp.Direction == 2].reset_index()
    for day in range(0, 112):
        count1 = 0
        count2 = 0
        for t in range(0, len(temp1.timestamp)):
            diff = (df.timestamp[i] - temp1.timestamp[t]).days
            if diff == day:
                count1 = count1 + temp1.Duration[t]
        for t in range(0, len(temp2.timestamp)):
            diff = (df.timestamp[i] - temp2.timestamp[t]).days
            if diff == day:
                count2 = count2 + temp2.Duration[t]
        values1.append(count1)
        values2.append(count2)
    array1.append(values1)
    array2.append(values2)
```

```python
newdf1 = pd.DataFrame(array1)
newdf2 = pd.DataFrame(array2)
```

```python
newdf1["timeseries"] = array1
newdf2["timeseries"] = array2
```

```python
df1 = pd.concat([df, newdf1], axis=1)
df2 = pd.concat([df, newdf2], axis=1)
```

```python
df1.to_csv("timeseriesCall1DepreST-CAT2021.csv", index = False)
df2.to_csv("timeseriesCall2DepreST-CAT2021.csv", index = False)
```
